```python
import math

import matplotlib.pyplot as plt


def plot_and_show(x_axis, y_axis, title):
    plt.plot(x_axis, y_axis)
    plt.title(title)
    plt.xlabel('Items')
    plt.ylabel('Operations')
    plt.show()


# Big O Notation
x = range(1, 1000)

## O(1) Constant Time
y = [1 for _ in x]
plot_and_show(x, y, "O(1)")
```

```python
## O(log n) Logarithmic Time
y = [math.log(i, 2) for i in x]
plot_and_show(x, y, "O(log n)")
```

```python
## O(n) Linear Time
y = [i for i in x]
plot_and_show(x, y, "O(n)")
```

```python
## O(n log n) Log-linear Time
y = [i * math.log(i, 2) for i in x]
plot_and_show(x, y, "O(n log n)")
```

```python
## O(n^2) Quadratic Time
y = [i ** 2 for i in x]
plot_and_show(x, y, "O(n^2)")
```

```python
## O(n^3) Cubic Time
y = [i ** 3 for i in x]
plot_and_show(x, y, "O(n^3)")
```

```python
## O(2^n) Exponential Time
x = range(1, 20)
y = [2 ** i for i in x]
plot_and_show(x, y, "O(2^n)")
```

```python
## O(n!) Factorial Time
x = range(1, 10)
y = [math.factorial(i) for i in x]
plot_and_show(x, y, "O(n!)")
```
